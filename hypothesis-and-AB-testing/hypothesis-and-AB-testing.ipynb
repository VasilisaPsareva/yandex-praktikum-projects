В первой части необходимо провести приоритизацию гипотез.

Во второй части мы выполним анализ результатов А/В тестирования.

## Описание данных

Данные для первой части:

* Файл /datasets/hypothesis.csv
    - Hypothesis — краткое описание гипотезы;
    - Reach — охват пользователей по 10-балльной шкале;
    - Impact — влияние на пользователей по 10-балльной шкале;
    - Confidence — уверенность в гипотезе по 10-балльной шкале;
    - Efforts — затраты ресурсов на проверку гипотезы по 10-балльной шкале. Чем больше значение Efforts, тем дороже проверка гипотезы.
    
Данные для второй части:

* Файл /datasets/orders.csv
    - transactionId — идентификатор заказа;
    - visitorId — идентификатор пользователя, совершившего заказ;
    - date — дата, когда был совершён заказ;
    - revenue — выручка заказа;
    - group — группа A/B-теста, в которую попал заказ.
    
* Файл /datasets/visitors.csv
    - date — дата;
    - group — группа A/B-теста;
    - visitors — количество пользователей в указанную дату в указанной группе A/B-теста

# Часть 1. Приоритизация гипотез.

## Знакомство с данными и предобработка

Импортируем необходимые билиотеки

```python
import pandas as pd
import scipy.stats as stats
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
```

Ознакомимся с таблицей `hypothesis`

```python
pd.set_option('max_colwidth', 400)
try:
    hypothesis = pd.read_csv('/datasets/hypothesis.csv')
except:
    hypothesis = pd.read_csv('datasets/hypothesis.csv')
hypothesis
```

```python
hypothesis.columns = hypothesis.columns.str.lower()
hypothesis
```

## ICE

Вспомним, как применяется метод рассчета приоритизации ICE:
    `ICE score = (Impact x Confidence) / Efforts`
Пользуясь данными датафрейма `hypothesis` посчитаем `ICE Score` для наших гипотез

```python
hypothesis['ICE'] = hypothesis['impact'] * hypothesis['confidence'] / hypothesis['efforts']

pd.set_option('display.float_format', '{:.2f}'.format)
hypothesis[['hypothesis', 'ICE']].sort_values(by='ICE', ascending=False)
```

Наиболее приоритетные гипотезы по методике ICE: 8, 0 и 7

## RICE

Вспомним, как применяется метод рассчета приоритизации RICE:
    `RICE score = (Reach x Impact x Confidence) / Efforts`
Пользуясь данными датафрейма `hypothesis` посчитаем `RICE Score` для наших гипотез

```python
hypothesis['RICE'] = hypothesis['reach'] * hypothesis['impact'] * hypothesis['confidence'] / hypothesis['efforts']

hypothesis[['hypothesis', 'RICE']].sort_values(by='RICE', ascending=False)
```

Наиболее приоритетные гипотезы по методике ICE: 7, 2 и 0

## Сравнение результатов приотиризации ICE и RICE

```python
hypothesis[['hypothesis', 'ICE', 'RICE']].sort_values(by='RICE', ascending=False)
```

В завсимости от используемого метода мы получаем несколько разную приотиризацию. На это влияет параметр `Reach`: гипотеза может быть сколь угодно простой в исполнении или оказывающей большое влияние, но если изменение затронет лишь малую долю пользователей ее реализация может быть не таким уж и выгодным.
Это мы видим например в 8 гипотезе: хоть она и лидирует по параметру ICE, ее охват минимален -- всего один балл из 10. С учетом охвата она оказывается не такой уж выгодной.

Суммируя результаты обеих методик можно сделать несколько выводов:

   1) 1, 3, 4 и 5 гипотезы обладают очень низким приоритетом.
   
   2) 8 гипотеза -- скидки в день рождения -- достаточно эффективна и проста, но не сильно повлияет на продажи за счет низкого охвата.
   
   3) 7 гипотеза -- самая приоритетная
   
   4) 2, 0 и 6 гипотезы так же обладают высокой приоритетностью.

## Предварительные рассчеты

Создадим функции для рассчета суммарных числа заказов, выручки и посетителей в зависимости от группы.

```python
#суммарная выручка и число заказов 

def get_orders_info(orders, group_name):
    orders_daily = (orders.query('group == @group_name')[['transaction_id', 'visitor_id', 'date', 'revenue']]
                    .groupby('date', as_index=False)
                    .agg({'transaction_id': pd.Series.nunique, 'revenue':'sum'}))
    orders_daily.columns = ['date', 'orders_per_date', 'revenue_per_date']
    orders_cummulative = (orders_daily.apply(lambda x: orders_daily[orders_daily['date'] <= x['date']]
                                           .agg({'date': 'max',
                                                 'orders_per_date': 'sum',
                                                 'revenue_per_date': 'sum'}),
                                           axis=1)
                                      .sort_values(by=['date']))

    orders_cummulative.columns = ['date', 'orders_cummulative', 'revenue_cummulative']
    return orders_daily, orders_cummulative
    
    
orders_daily_A, orders_cummulative_A = get_orders_info(orders, 'A')
orders_daily_B, orders_cummulative_B = get_orders_info(orders, 'B')
```

```python
#суммарное число посетителей
def get_cummulative_visits(visitors, group_name):
    visitors_daily = visitors.query('group == @group_name').reset_index()[['date', 'visitors']]
    visitors_daily.columns = ['date', 'visitors_per_date']
    visitors_cummulative = (visitors_daily.apply(lambda x: visitors_daily[visitors_daily['date'] <= x['date']]
                                                        .agg({'date': 'max', 'visitors_per_date': 'sum'}),
                                                axis=1).sort_values(by=['date']))
    visitors_cummulative.columns = ['date', 'visitors_cummulative']
    return visitors_daily, visitors_cummulative

visitors_daily_A, visitors_cummulative_A = get_cummulative_visits(visitors, 'A')
visitors_daily_B, visitors_cummulative_B = get_cummulative_visits(visitors, 'B')
```

## Анализ графиков показателей групп по сырым данным

### График кумулятивной выручки по группам

```python
fig, ax = plt.subplots(figsize=(12, 6))

plt.plot(orders_cummulative_A['date'], orders_cummulative_A['revenue_cummulative'], label='A')
plt.plot(orders_cummulative_B['date'], orders_cummulative_B['revenue_cummulative'], label='B')
plt.legend()
plt.tick_params(labelrotation=45)
plt.ticklabel_format(style='plain', axis='y')
plt.grid()
```

Суммарная выручка группы B практически сразу выше чем для группы А. Около 18 августа происходит скачок выручки группы В -- возможно тогда был совершен аномально дорогой заказ, который нам не стоит учитывать.

### График кумулятивного среднего чека по группам

```python
fig, ax = plt.subplots(figsize=(12, 6))

plt.plot(orders_cummulative_A['date'],
         orders_cummulative_A['revenue_cummulative']/orders_cummulative_A['orders_cummulative'],
         label='A')
plt.plot(orders_cummulative_B['date'],
         orders_cummulative_B['revenue_cummulative']/orders_cummulative_B['orders_cummulative'],
         label='B')
plt.legend()
plt.tick_params(labelrotation=45)
plt.ticklabel_format(style='plain', axis='y')
plt.grid()
```

Кумулятивный средний чек обеих групп сильно колеблется примерно до 15.08. 18.08 мы видим большой скачок для группы В, там же где видели скачок выручки. Это косвенно подтверждает предположение об аномально дорогом заказе в эту дату.
После скачка средний чек группы В плавно падает, в то время как для группы А значение плавно растет.

### График относительного изменения кумулятивного среднего чека группы B к группе A

```python
fig, ax = plt.subplots(figsize=(12, 6))

plt.plot(orders_cummulative_A['date'],
         ((orders_cummulative_B['revenue_cummulative']/orders_cummulative_B['orders_cummulative'])/
          (orders_cummulative_A['revenue_cummulative']/orders_cummulative_A['orders_cummulative']) - 1))

plt.tick_params(labelrotation=45)
plt.axhline(y=0, color='black', linestyle='--')
plt.ticklabel_format(style='plain', axis='y')
plt.grid()
```

На этом графике мы видим много скачков: сначала группа В по среднему чеку значительно проигрывала, потом резко выигрывала, около 08.08 снова начала проигрывать, а 18.08 -- относительное изменение резко выросло, и дальше плавно падало. По этому графику сложно сделать уверенные выводы о выгодности гипотезы.

После начальных колебаний можно достаточно уверенно заявить, что для группы В конверсия стабильно держится выше чем для группы А, но различие всегго несколько процентов.

### Рассчеты по сырым данным

Напишем функцию, создающую выборки количества заказов по пользователям

```python
def get_user_info(orders, visitors_daily, group_name):
    orders_by_user = (orders.query('group == @group_name')
                     .groupby('visitor_id', as_index=False)
                     .agg({'transaction_id': pd.Series.nunique})
                     )
    orders_by_user.columns = ['user', 'orders']
    sample = pd.concat([orders_by_user['orders'],
                       pd.Series(0, index=np.arange(visitors_daily['visitors_per_date'].sum() 
                                                    - len(orders_by_user['orders']))
                                , name='orders')
                       ],
                       axis=0)
    return sample
```

Посчитаем статистическую значимость различия выборок количества заказов по пользователям для разных групп. Сформулируем гипотезы так:

    H0: конверсии групп А и В равны
    H1: конверсии групп А и В различаются

```python
sample_A = get_user_info(orders, visitors_daily_A, 'A')
sample_B = get_user_info(orders, visitors_daily_B, 'B')

results = stats.mannwhitneyu(sample_A, sample_B)
alpha = .05 # критический уровень статистической значимости results = st.mannwhitneyu(data1, data2)
print('p-значение: ', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу: разница статистически значима")
else:
    print("Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя")


print("{0:.5f}".format(sample_B.mean() / sample_A.mean() - 1))
```

p-value <  𝛼, нулевая гипотеза опровергнута, значит конверсии можно считать разными. При этом относительный выигрыш конверсии для группы В -- 13.8%.

Посчитаем статистическую значимость различий в среднем чеке заказа между группами:

    H0: средние чеки групп А и В равны
    H1: средние чеки групп А и В различаются

```python
results = stats.mannwhitneyu(orders[orders['group']=='A']['revenue'], orders[orders['group']=='B']['revenue'])
alpha = .05 # критический уровень статистической значимости results = st.mannwhitneyu(data1, data2)
print('p-значение: ', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу: разница статистически значима")
else:
    print("Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя")


print('{0:.3f}'.format(orders[orders['group']=='B']['revenue'].mean()/
                       orders[orders['group']=='A']['revenue'].mean() - 1)) 
```

p-value около 0.73. Это высокая статистическая значимость. Значит нельзя сделать вывод о различии выручки между групп. В то же время относительная разница выручки около 26%. Это может быть результатом аномалий -- при проверке критерием Манна-Уитни выбросы не влияют на результат. А на рассчет различия они могут влиять.

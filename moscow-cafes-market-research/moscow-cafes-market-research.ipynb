### Анализ распределения посадочных мест

Проанализируем что более характерно для сетевых заведений: много заведений с небольшим числом посадочных мест в каждом или мало заведений с большим количеством посадочных мест?

Для начала определим, сколько заведений мы будем считать за "много", какое количество посадочных мест -- "большое"? Для этого найдем все сети, количество заведений в них и среднее количество посадочных мест.

```python
data_by_name = (data.query('chain=="да"')
     .groupby('object_name', as_index=False)
     .agg({'id': 'count', 'number': 'mean'})
     )
data_by_name.columns = ['object_name', 'count', 'avg_sits']
data_by_name = data_by_name.query('count > 1') #отбросим сети с одним заведением -- это скорее ошибка в данных
                                            # или в названии заведения
```

#### Для количества заведений:

```python
np.percentile(data_by_name['count'], [70, 90, 95])
```

Добавим категории количества заведений: до 6(включительно) заведений -- малое количество заведений, выше -- большое

```python
def count_of_places(numb):
    if numb > 6:
        return 1
    else:
        return 0
    
data_by_name['places_cat'] = data_by_name['count'].apply(count_of_places)
```

#### Для количества посадочных мест:

```python
np.percentile(data_by_name['avg_sits'], [70, 90, 95])
```

Будем считать, что более 70 -- много посадочных мест. Категоризируем данные в 4 категории:

    * мало посадочных мест и мало заведений
    * много посадочных мест и мало заведений
    * мало посадочных мест и много заведений
    * много посадочных мест и много заведений

```python
def categorize(row):
    if row['avg_sits'] > 70:
        return row['places_cat']*2 + 1
    else:
        return row['places_cat']*2
    
def name_category(number):
    if number == 0:
        return "мало посадочных мест и заведений"
    elif number == 1:
        return "много посадочных мест и мало заведений"
    elif number == 2:
        return "мало посадочных мест и много заведений"
    else:
        return "много посадочных мест и заведений"
    
    
data_by_name['category'] = data_by_name.apply(categorize, axis=1)
data_by_name['category'] = data_by_name['category'].apply(name_category)
data_by_name.groupby('category').agg({'object_name': 'count'}).sort_values(by='object_name')
```

```python
ax = sns.jointplot(
    data=data_by_name,
    x='count',
    y='avg_sits',
    hue='category',
    height=8,
    palette=sns.husl_palette(4, l=0.7)
)
ax.ax_joint.set_xlabel('Количество точек в сети')
ax.ax_joint.set_ylabel('Среднее количество посадочных мест в сети');
```

Больше всего сетей с малым количеством точек и посадочных мест, а вот сетей с большим количеством точек и посадочных мест -- почти нет.
Сетей с большим количеством посадочных мест и малым количеством точек чуть больше, чем сетей в которых напротив, много точек и мало посадочных мест.

Для каждого вида объекта общественного питания найдем среднее количество посадочных мест: 

```python
ordered = data.groupby('object_type', as_index=False).mean().sort_values(by='number', ascending=False)

fig = plt.figure(figsize=(12, 8))

sns.barplot(data=data,
            x='object_type',
            y='number',
            estimator=np.mean,
            order=ordered['object_type']
);

plt.title("Среднее количество посадочных мест для разных видов заведений")
plt.xlabel("Вид заведения")
plt.ylabel("Среднее количество посадочных мест");
```

Самые массовые заведения -- столовые и рестораны.

#### Вывод

Чаще всего встречаются сети в которых мало точек и мало посадочных мест. Так же немало сетей, где много посадочных мест и мало заведений, как и наоборот.

В целом больше всего посадочных мест в столовых и ресторанах.

### Расположение объектов питания

#### Выделение улицы из адреса

```python
address_apart = data['address'].str.split(', ')
address_apart # адреса разбили на списки строк
```

```python
def get_street(address_row):
    prev_word=address_row[0]
    for word in address_row:
        if ('дом' in word 
            or 'километр' in word       #улица чаще всего идет перед номером дома
            or 'владение' in word):
            return prev_word
        prev_word=word
        
data['street'] = address_apart.apply(get_street)
address_apart[address_apart.apply(get_street).isna()]
```

Практически для всех адресов улицы выделили, остались лишь сотня адресов, в которых улица скорее всего не обозначена.
В нашем методе есть минусы, наверняка не все названия корректны, но в основной массе улицы выделены.

#### Топ-10 улиц по количеству объектов

Выделим 10 улиц с самым большим количеством объектов на них:

А кулинарии практически совсем без посадочных мест.

Наша цель -- небольшое кафе. Столовая, ресторан, кулинария -- точно не наш формат. Как и кафетерий с буфетом -- там самообслуживание, и в роботах не будет нужды.

```python
fig = plt.figure(figsize=(12, 8))

our_types=['кафе', 'закусочная', 'фастфуд', 'ресторан', 'бар']
ax = sns.violinplot(data=data.query('object_type in @our_types'),
                     y='number', x='object_type')
ax.set_aspect('auto')
ax.set_ybound(lower=-50, upper=300)
plt.title('Посадочные места для разных видов объектов')
plt.xlabel('Вид объекта')
plt.ylabel('Распределение посадочных мест');
```

Теперь мы видим конкретные различия и можем предложить несколько вариантов заведений:

    * закусочная до 20 посадочных мест
